```python
import pandas as pd
nd=pd.read_excel("/content/cricket data.xlsx")
nd
```

```python
print("Shape:",nd.shape)
print("\nInfo:")
print(nd.info())
print("\nStatistics:")
print(nd.describe)
```

```python
print(nd.isnull().sum())
invalid_value=["NA","Unkown","","---","-"]
print(nd.isin(invalid_value).sum())
```

```python
import numpy as np
nd.replace(["NA", "Unknown", "", "---", "-"], np.nan, inplace=True)
nd
```

```python
nd["Runs_Scored"] = pd.to_numeric(nd["Runs_Scored"], errors="coerce")

nd["Balls_Faced"] = pd.to_numeric(nd["Balls_Faced"], errors="coerce")

nd["Strike_Rate_Calc"] = pd.to_numeric(nd["Strike_Rate_Calc"], errors="coerce")

nd["Overs_Numeric"] = pd.to_numeric(nd["Overs_Numeric"], errors="coerce")

nd["Runs_Scored"] = pd.to_numeric(nd["Runs_Scored"], errors="coerce")

nd["Wickets_Taken"] = pd.to_numeric(nd["Wickets_Taken"], errors="coerce")
nd
```

```python
nd["Batting_StrikeRate"] = (nd["Runs_Scored"] / nd["Balls_Faced"]) * 100
nd
```

```python
nd["Bowling_Economy"] = nd["Runs_Scored"] / nd["Overs_Numeric"]
nd
```

```python
venue_summary = nd.groupby("Venue").agg({
    "Runs_Scored": "mean",
    "Batting_StrikeRate": "mean"
}).reset_index()

print(venue_summary)
```

```python
batsman_summary = nd.groupby("Batsman_Name").agg({
    "Runs_Scored": "sum",
    "Wickets_Taken": "mean"
}).reset_index()

print(batsman_summary)
```

```python
import pandas as pd

# 1. Aggregate runs scored by Batting_Team for each Match_ID
#    This sums up all individual batsman runs for a given team in a given match.
team_total_runs = nd.groupby(['Match_ID', 'Batting_Team'])['Runs_Scored'].sum().reset_index()

# 2. Determine match outcomes for each match
match_outcomes = {}
for match_id in team_total_runs['Match_ID'].unique():
    match_teams_data = team_total_runs[team_total_runs['Match_ID'] == match_id]

    if len(match_teams_data) == 2: # Expecting two teams per match for a clear win/loss
        team1_name = match_teams_data.iloc[0]['Batting_Team']
        team1_score = match_teams_data.iloc[0]['Runs_Scored']
        team2_name = match_teams_data.iloc[1]['Batting_Team']
        team2_score = match_teams_data.iloc[1]['Runs_Scored']

        if team1_score > team2_score:
            match_outcomes[match_id] = {team1_name: 'Win', team2_name: 'Loss'}
        elif team2_score > team1_score:
            match_outcomes[match_id] = {team1_name: 'Loss', team2_name: 'Win'}
        else: # Tie or Draw
            match_outcomes[match_id] = {team1_name: 'Tie/Draw', team2_name: 'Tie/Draw'}
    elif len(match_teams_data) < 2:
        # Handle cases where only one team's data is available (e.g., incomplete match, data error)
        for _, row in match_teams_data.iterrows():
            match_outcomes.setdefault(match_id, {})[row['Batting_Team']] = 'Incomplete'
    else:
        # Handle cases where more than two teams appear for a Match_ID (e.g., data error or specific formats like triangular series where one match_id might involve more than 2 teams)
        # For simplicity, these will be marked as 'Ambiguous' or 'Unknown' as direct win/loss comparison is difficult.
        for _, row in match_teams_data.iterrows():
            match_outcomes.setdefault(match_id, {})[row['Batting_Team']] = 'Ambiguous'

# 3. Update the 'Match_Result' column in the original DataFrame `nd`
def assign_match_result(row):
    match_id = row['Match_ID']
    batting_team = row['Batting_Team']
    return match_outcomes.get(match_id, {}).get(batting_team, 'Unknown')

nd['Match_Result'] = nd.apply(assign_match_result, axis=1)
display(nd.head())
```

```python
top_performances = nd[nd["Match_Result"] == "Win"] \
                    .sort_values(by="Runs_Scored", ascending=False) \
                    .head(10)

print(top_performances)
nd
```
